```python
import requests
from bs4 import BeautifulSoup
```

```python
r=requests.get("https://www.w3schools.com/html/html_examples.asp")
c=r.content
```

```python
c
```

```python
soup=BeautifulSoup(c,"html.parser")
```

```python
all = soup.find_all("div",{"class":"w3-bar-block"})
```

```python
all[0]
```

```python
all[0].find_all("h3")[0].text
```

```python
for item in all:
    print(item.find_all("a")[0].text)
```
